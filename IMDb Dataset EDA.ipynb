```python
# ...existing code...
# Create the dataframe df_by_genre
df_by_genre = movies.loc[:, 'CVotes10':'VotesnUS'].copy()
df_by_genre['genre_1'] = movies['genre_1']
df_by_genre['genre_2'] = movies['genre_2']
df_by_genre['genre_3'] = movies['genre_3']
df_by_genre['cnt'] = 1 

# Now group only by the genre columns, which are not included in the sum
df_by_g1 = df_by_genre.groupby('genre_1').sum(numeric_only=True)
df_by_g2 = df_by_genre.groupby('genre_2').sum(numeric_only=True)
df_by_g3 = df_by_genre.groupby('genre_3').sum(numeric_only=True)

# Add the grouped data frames and store it in a new data frame
df_add = df_by_g1.add(df_by_g2, fill_value=0)
df_add = df_add.add(df_by_g3, fill_value=0)
df_add
# ...existing code...
```

```python
# Extract genres with atleast 10 occurences

genre_top_10 = df_add.loc[df_add.cnt >10]
genre_top_10
```

```python
# Take the mean for every column by dividing with cnt 

# Divide all columns except 'cnt' by 'cnt' to get the mean
genre_top_10 = genre_top_10.div(genre_top_10['cnt'], axis=0)
genre_top_10['cnt'] = genre_top_10['cnt'] # keep cnt as is (or you can drop it if not needed)
genre_top_10
```

```python
# Rounding off the columns of Votes to two decimals

# Find all columns that start with 'Votes'
votes_col = [col for col in genre_top_10.columns if col.startswith('Votes')]

# Round the specified columns to 2 decimal places
genre_top_10[votes_col] = genre_top_10[votes_col].round(2)
genre_top_10
```

```python
# Converting CVotes to int type

# Find all columns that start with 'CVotes'
cvotes_cols = [col for col in genre_top_10.columns if col.startswith('CVotes')]

# Convert these columns to integer type
genre_top_10[cvotes_cols] = genre_top_10[cvotes_cols].astype(int)
genre_top_10
```
